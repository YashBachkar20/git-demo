# Plot a bar graph for Host_ID vs Count of Listings.

```python
grouped = NYC.groupby('host_id').count()['id']
grouped = grouped.sort_values(ascending = False)
N = 50
grouped = grouped[:N]


grouped.plot(kind = 'bar', figsize = (15, 8))

plt.title("Count of Listings by Host ID")
plt.xlabel("Host ID")
plt.ylabel("Count of Listings")

plt.show()
```

From bar graph we the Host_id 219517861 with maximum listing count with more than 300 count

and minimum count of less than 50 with Host-id 260639745

# Based on Room_Type, prepare a bar graph for neighbourhood vs Count of Listings

```python
nbh = NYC.loc[NYC['neighbourhood'].isin(['Williamsburg','Bedford-Stuyvesant','Harlem','Bushwick','Upper West Side','Hell\'s Kitchen','East Village','Upper East Side','Crown Heights','Midtown'])]

grouped = nbh.groupby(['room_type', 'neighbourhood']).count()['host_id']
grouped = grouped.reset_index()
pivot = grouped.pivot(index = 'neighbourhood', columns = 'room_type', values = 'host_id')
print(pivot)
pivot.plot(kind = 'bar', figsize = (15, 8))

plt.title("Count of Listings by Room Type and Neighbourhood")
plt.xlabel("Neighbourhood")
plt.ylabel("Count of Listings")

plt.show()
```

We seen that,

Bedford-stuyvesant and Williamsburg have max listing count for Private Room type
